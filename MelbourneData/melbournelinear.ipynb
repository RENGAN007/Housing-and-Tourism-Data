```python
import numpy as np
import pandas as pd
import sklearn 
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Melbourne_housing_FULL.csv")
df.to_string()
df
```

```python
col = ['Suburb','Rooms','Type','Price','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea']
df=df[col]
df
```

```python
X=df.dropna(inplace=True)
X=pd.get_dummies(df,dtype='int')
X=X.drop(columns=['Price'])
Y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
```

```python
model.score(x_train.values,y_train.values)
```

```python
model.score(x_test.values,y_test.values)
```

```python
from sklearn.linear_model import Lasso
lassobro = Lasso(alpha=50,max_iter=50,tol=0.2)
lassobro.fit(x_train,y_train)
```

```python
lassobro.score(x_train.values,y_train.values)
```

```python
lassobro.score(x_test.values,y_test.values)
```

```python

from sklearn.linear_model import Ridge
lassobro=Ridge(alpha=50,max_iter=100,tol=0.1)
lassobro.fit(x_train,y_train)
```

```python
lassobro.score(x_test,y_test)
```

```python
lassobro.score(x_train,y_train)
```
