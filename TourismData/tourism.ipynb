```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
```

```python
tour=pd.read_csv('Top Indian Places to Visit.csv')
tour
```

```python
coll=['Type','Google review rating','Number of google review in lakhs','Entrance Fee in INR']
tours=tour[coll]
tours
```

```python
tours = tours[tours['Entrance Fee in INR'] != 0]
tours
```

```python
tours=pd.get_dummies(tours,dtype='int')
x=tours.drop(columns=['Entrance Fee in INR'])
y=tours['Entrance Fee in INR']
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35)
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.linear_model import Lasso
ttl=Lasso(alpha=50,max_iter=100,tol=0.1)
ttl.fit(x_train,y_train)
```

```python
ttl.score(x_train,y_train)

```

```python
from sklearn.linear_model import Ridge
model3=Ridge(alpha=50,max_iter=100,tol=0.1)
model3.fit(x_train,y_train)
```

```python
model3.score(x_test,y_test)
```

```python
model3.score(x_train,y_train)
```
